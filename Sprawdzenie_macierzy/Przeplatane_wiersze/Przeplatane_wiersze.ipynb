```python
##############################################################
##### Wczytanie zbioru testowego po rozszerzeniu danych  #####
##############################################################
```

```python
#Wczytanie etykiet do testowania

with open(  '../../../Dane/Baza_zdjec_przeplatane_2/Test_2/test_2.csv'   ,"r") as csvfile:
    lines = csvfile.readlines()
csvfile.close()

new = []

for i in lines:
    new.append(i.split(" "))

    
value_test = []
for i in new[0]:
    value_test.append(int(i))    
    
print(len(value_test))
```

```python
#Wczytanie zdjęć testowych 
images_test = []
size = len(value_test)

for i in range(size):
    im = Image.open('../../../Dane/Baza_zdjec_przeplatane_2/Test_2/'+str(i)+'.png')
    arr_im = np.array(im) 
    arr_im = arr_im / 255.0
    images_test.append(arr_im) 
print(len(images_test))
```

```python
x_test = np.array(images_test)
y_test = np.array(value_test)
```

```python
a = 2
plt.imshow(x_test[a],cmap = 'gray')
print(y_test[a])
```

```python
best_model = tf.keras.models.load_model('Model_przeplatane_wiersze.h5', compile=False) 

best_model.compile(optimizer='adam',
              loss='mse',
              metrics=['accuracy','mse'])
```

```python
best_model.evaluate(x_test,  y_test, verbose=2)
```

```python
predict_x = best_model.predict(x_test) 

predictions = (predict_x > 0.5).astype("int32")
```

```python
names = [['Tn','Nn'],['Np', 'Tp']]
names = np.reshape(names,(2,2))
print(names)

conf = tf.math.confusion_matrix(y_test,predictions)

print(conf)
```

```python
predictions.tofile('predictions.csv', sep = '\n') 
predict_x.tofile('predict_x.csv', sep = '\n') 
y_test.tofile('test_label.csv', sep = '\n') 
```
