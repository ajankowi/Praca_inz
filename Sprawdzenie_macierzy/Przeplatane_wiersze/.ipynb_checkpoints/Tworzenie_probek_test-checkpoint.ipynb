```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from PIL import Image
import os
import csv
import random

import cv2
import imutils

# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
```

```python
#Sciezki do folderow
folder = '../Baza_zdjec/s'
folder_docelowy = './Dane_testowe/'
```

```python
test_ppl = []    
for i in range(37,41):
    images = []
    for j in range(1,11):
        im = Image.open(folder+str(i)+'/'+str(j)+'.png')
        arr_im = np.array(im) 
        images.append(arr_im) 
    test_ppl.append(images) 
```

```python
w, h = 64, 128
data = np.zeros((h, w), dtype=np.uint8)
```

```python
ilosc = 40
rozmiar = 10
test_img = []
n = 0

# Inicjalizacja tablicy zerami o rozmiarze 'rozmiar'
tab_test = np.zeros(360, dtype=np.uint8)


#Stworzenie nowego katalogu
newpath = folder_docelowy
#if not os.path.exists(newpath):
#    os.makedirs(newpath)

#Petla przechodzi po ludziach
for i in range(4):
    
    #Petla przechodzi po zdjeciach
    for l in range(1):
    
        m = 10
    
        for j in range(rozmiar):       
            m = m - 1
        
            for k in range(m):

                a = 0
                
                #Pętla służy do przeplatania wierszy zdjęć
                for p in range(len(test_ppl[i][j+l])):
                    p = 2*p
                    data[p] = test_ppl[i][j+l][a]
                    data[p+1] = test_ppl[i][9-k+l][a]
                    a = a + 1      
    
                #Zapisanie zmienionego zdjecia
                con = Image.fromarray(data)
                con.save(folder_docelowy+str(n)+'.png') 
      
                test_img.append(con)
                tab_test[n] = 1 
                n = n + 1
                
print(n)
```

```python
x = 0 #Zamiast n 

polaczone = n
m = 5
rozne_test = []


for i in range(4):

    rand_1 = random.randint(1,4)
    rand_2 = random.randint(1,4)
   
    
    m = m - 1
    
    for k in range(1,m):
    
        for j in range(5):
       
            for l in range(5):
            
                a = 0
                data = np.zeros((h, w), dtype=np.uint8)
                
                #Pętla służy do przeplatania wierszy zdjęć
                for p in range(w):
                    p = 2*p
                    data[p] = test_ppl[i][rand_1+j][a]
                    data[p+1] = test_ppl[i+k][rand_2+l][a]
                    a = a + 1
                
                
                if n < 360:
                
                
                    con = Image.fromarray(data)
                    x = x + 1
                    rozne_test.append(con)
            
                    con.save(newpath+str(n)+'.png')
                
                    tab_test[n] = 0
                    n = n + 1
            

                   
print(x)
print(n)
```

```python
#Zapis tablicy do pliku CSV
tab_test.tofile(newpath+'test_label_clear.csv', sep = ' ') 
```
