```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from PIL import Image
import os

# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
```

```python
folder = 'Baza_zdjec/s'
folder_docelowy = 'Baza_zdjec_przerobiona/s'
slesh = '/'
pgm = '.pgm'
png = '.png'
```

```python
#Wlasciwa czesc kodu
w, h = 92, 92

for i in range(1,41):
    for j in range(1,11):
        im = Image.open(folder+str(i)+slesh+str(j)+pgm)
        im_arr = np.array(im)
        
        #Tworzenie pustej tablicy
        data = np.zeros((h, w), dtype=np.uint8)
        data[0:w, 0:w] = [255]
        
        #Kopiowanie 92 wierszy (kwadratowy obraz)
        for x in range(w):
            data[x] = im_arr[10+x]
        
        #konwersja tablicy znak√≥w na obraz
        con = Image.fromarray(data)

        #Zmiana rozmiaru zdjec
        con_2 = con.resize((int(64), int(64)))
        
        #Stworzenie nowego katalogu
        newpath = folder_docelowy+str(i)
        if not os.path.exists(newpath):
            os.makedirs(newpath)
            
        #Zapisanie zmienionego zdjecia
        con_2.save(newpath+slesh+str(j)+png)        
```

```python

#Stare

newpath = r'Baza_zdjec_przerobiona/s4' 
if not os.path.exists(newpath):
    os.makedirs(newpath)
else:
    print("Juz istnieje folder!")
con_2.save(r'Baza_zdjec_przerobiona/s4/1.png')
```
