```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from PIL import Image
import os

import cv2
import imutils

# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
```

```python
folder = 'Baza_zdjec/s'
folder_docelowy = 'Baza_zdjec_przerobiona/s'
slesh = '/'
png = '.png'
```

```python
images = []
people = []

for i in range(1,9):
    n = i * 5
    images = []
    for j in range(0,4):
        k = 4 + j * 40
        im = Image.open(folder_docelowy+str(n)+slesh+str(k)+png)
        images.append(im) 
    people.append(images)
```

```python
f, axarr = plt.subplots(2,4)
for i in range(0,2):
    for j in range(0,4):
        axarr[i,j].imshow(people[i][j], cmap = 'gray')
```

```python
f, axarr = plt.subplots(2,4)
for i in range(0,2):
    for j in range(0,4):
        axarr[i,j].imshow(people[i+2][j], cmap = 'gray')
```
