```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from PIL import Image
import os

import cv2
import imutils


# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
```

```python
#im = Image.open('jablko.jfif')

im = cv2.imread('jablko.jfif')

im = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

#im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(im, cmap = 'gray')
arr_im = np.array(im)

print(im.shape)
```

```python
rot = imutils.rotate(im, angle=-5)
```

```python
plt.imshow(rot, cmap = 'gray')
```

```python
im_2 = im/255

mean = 0
var = 0.01
sigma = np.sqrt(var)


n = np.random.normal(mean,sigma,size=(225,225))

g = im_2 + n
```

```python
g = g * 255
plt.imshow(g, cmap = 'gray')
```

```python
# Mirror in x direction (flip horizontally)
imgX = np.flip(im, axis=1)

# Mirror in y direction (flip vertically)
imgY = np.flip(im, axis=0)
```

```python
plt.imshow(imgX, cmap = 'gray')
```

```python
plt.imshow(imgY, cmap = 'gray')
```
